# **IMPORTS + DATA**

```python
import numpy as np
import torch
```

```python
inputs=np.array([[73.,67,45],[91,88,64],[87,134,58],[102,43,37],[69,96,70]])
```

```python
inputs
```

```python
target=np.array([[56.,70],[81,101],[119,113],[22,37],[103,119]])
```

```python
target
```

```python
inputs = torch.from_numpy(inputs).to(torch.float32)
inputs
target = torch.from_numpy(target).to(torch.float32)
target
```

# **MODEL AND LOSS**

```python
#MODEL PARAMETERS
```

```python
w=torch.randn(2,3,requires_grad=True)
b=torch.randn(2,requires_grad=True)
print(w)
print(b)
```

```python
def model(x):
  return x@ w.t() +b
```

```python
inputs.float() @ w.t() +b
```

```python
def mse(t1,t2):
  diff=t1-t2
  return torch.sum(diff*diff)/diff.numel()
```

# **FORWARD AND BACKWARD PASS**

```python
preds=model(inputs.float())
preds
```

```python
diff=preds-target
diff
```

```python
mse(preds,target)
```

```python
loss=mse(preds,target)
```

```python
loss.backward()
```

```python
print(w.grad)
```

```python
print(loss)
```

```python
print(b.grad)
```

```python
pred=model(inputs.float())
```

```python
with torch.no_grad():
  w-=w.grad*1e-5
  b-=b.grad*1e-5
```

```python
loss=mse(preds,target)
```

```python
loss
```

# **TRAINING LOOP**

```python

learning_rate = 1e-5
epochs = 100

for epoch in range(epochs):
    preds = model(inputs)
    loss = mse(preds, target)

    loss.backward()

    with torch.no_grad():
        w -= learning_rate * w.grad
        b -= learning_rate * b.grad

        w.grad.zero_()
        b.grad.zero_()

    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item():.4f}")
```
