# **TITANIC DATA SURVIVAL PREDICTION**

```python
import torch.nn as nn
import torch.optim as optim
import pandas as pd
import numpy as np
from torch.utils.data import DataLoader,TensorDataset
import torch
```

# **LOAD DATA**

```python
titanic=pd.read_csv('/Titanic.csv')
```

```python
titanic.head()
```

```python
titanic.columns
```

```python
titanic.shape
```

# **DATA PREPROCESSING**

# NULL VALUE HANDLING

```python
(titanic.isnull()==True).sum()
```

```python
titanic["age"].fillna(titanic["age"].mean(),inplace=True)
```

```python
(titanic.isnull()==True).sum()
```

```python
titanic['embarked'].fillna(titanic['embarked'].mode()[0], inplace=True)
```

```python
titanic.isnull().sum()
```

```python
titanic.drop(columns=["who"],inplace=True)
```

# ENCODE STRING VALUES

```python
titanic.columns
```

```python
titanic=pd.get_dummies(titanic,columns=['sex','embarked','alone','class'])
```

```python
titanic.head()
```

```python
print(titanic.columns)
```

```python
titanic = titanic.astype(int)
```

```python
titanic.head()
```

# HANDLE DUPLICATES

```python
titanic.drop(columns=["sex_male","class_Third"],inplace=True)
```

```python
titanic.head()
```

```python
titanic.drop(columns=["embarked_S","alone_False"],inplace=True)
```

```python
titanic.head()
```

# **SETTING TARGET AND INPUT**

```python
x=titanic.drop(columns=["survived"])
```

```python
x
```

```python
y=titanic["survived"]
```

```python
x=torch.tensor(x.values,dtype=torch.float32)
y=torch.tensor(y.values,dtype=torch.long)
```

```python
x = (x - x.mean(dim=0)) / x.std(dim=0)
```

```python
x.shape,y.shape
```

```python
dataset=TensorDataset(x,y)
```

```python
dataloader=DataLoader(dataset=dataset,shuffle=True,batch_size=64)
```

```python
input_dim=x.shape[1]
hidden_dim=24
output_dim=2
```

# **MODEL CREATION**

```python
model=nn.Sequential(
    nn.Linear(in_features=input_dim,out_features=hidden_dim),
    nn.ReLU(),
    nn.Linear(in_features=hidden_dim,out_features=output_dim)
)
```

```python
optimizer=optim.Adam(model.parameters(),lr=0.005)
loss_fn = nn.CrossEntropyLoss()
```

# **TRAINING LOOP**

```python
epochs=200
for epoch in range(epochs):
  for i,(inputs,labels) in enumerate(dataloader):
    outputs=model(inputs)
    loss=loss_fn(outputs,labels)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
  if epoch%10==0:
    print(f"Epoch {epoch+1}/{epochs}, Loss: {loss.item()}")
```

# **MODEL EVALUATION**

```python
correct = 0
total = 0

with torch.no_grad():
    for inputs, labels in dataloader:
        outputs = model(inputs)
        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print(f"Accuracy: {100 * correct / total:.2f}%")
```
